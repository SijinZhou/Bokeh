```python
import pandas as pd
from bokeh.palettes import Spectral4
from bokeh.plotting import figure, output_file, show
```

```python
# import sample data that only contain records from 4 universities of original data
dataset=pd.read_excel('admission_v2.xlsx')
print(dataset.head())
```

```python
#select columns we need
df = pd.DataFrame(dataset, columns=['Year','Institution Name','Admissions Total'])
```

```python
df
```

```python
df_1 = df[df['Institution Name']=='Abilene Christian University']
df_2 = df[df['Institution Name']=='Abraham Baldwin Agricultural College']
df_3 = df[df['Institution Name']=='Adams State University']
df_4 = df[df['Institution Name']=='Adelphi University']
```

```python
df_1
```

```python
dict1 = df_1.to_dict(orient = 'list')
dict2 = df_2.to_dict(orient = 'list')
dict3 = df_3.to_dict(orient = 'list')
dict4 = df_4.to_dict(orient = 'list')
```

```python
dict2
```

```python
#set up the layout 
p = figure(plot_width=800, plot_height=250, x_axis_type="datetime")
p.title.text = 'Click on legend entries to hide the corresponding lines'
```

```python
for data, name, color in zip([dict1,dict2,dict3,dict4], ["Abilene Christian University", "Abraham Baldwin Agricultural College", "Adams State University", "Adelphi University"], Spectral4):
    newdf = pd.DataFrame(data)
    newdf['Year'] = pd.to_datetime(newdf['Year'])
    p.line(newdf['Year'], newdf['Admissions Total'], line_width=2, color=color, alpha=0.8, legend_label=name)
```

```python
#set up lengend out put and output plot as a html
p.legend.location = "top_left"
p.legend.click_policy="hide"

output_file("interactive_legend.html", title="interactive_legend.py example")

show(p)
```
