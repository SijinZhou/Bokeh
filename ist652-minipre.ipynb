```python

import pandas as pd
```

```python
from bokeh.io import output_file,show,output_notebook; output_notebook()
```

```python

from bokeh.models import (BasicTicker, ColorBar, ColumnDataSource,
                          LinearColorMapper, PrintfTickFormatter,)
from bokeh.models import HoverTool
from bokeh.plotting import figure
from bokeh.transform import transform
```

```python
df=pd.read_excel(r'University-Admissions-Data-2009-2018.xlsx')
```

```python
df.head()
```

```python
data=pd.DataFrame()
```

```python
data['Year'] =pd.DatetimeIndex(df['Year']).year
```

```python
data['University']=df['Institution Name']
```

```python
data['Admissions']=df['Admissions Total']

#Admission_score=(data.Admissions-data.Admissions.mean())/(data.Admissions.max()-data.Admissions.min())*100
#data['Admission_score']=Admission_score
```

```python
data=data.dropna()
data.Year = data.Year.astype(str)
data = data.set_index('Year')
```

```python
data
```

```python
source = ColumnDataSource(data)  

colors = ["#75968f", "#a5bab7", "#c9d9d3", "#e2e2e2", "#dfccce", "#ddb7b1", "#cc7878", "#933b41", "#550b1d"]
mapper = LinearColorMapper(palette=colors, low=data.Admissions.min(), high=data.Admissions.max())

p = figure(plot_width=10000, plot_height=500,title="University-Admissions-2009-2018",
            y_range=list(data.index.unique()), x_range=list(data.University.unique()),
           toolbar_location=None, tools="hover", x_axis_location="above")

p.rect(x="University", y="Year", width=1, height=1, source=source,
       line_color=None, fill_color=transform('Admissions', mapper))

color_bar = ColorBar(color_mapper=mapper, location=(0, 0),
                     ticker=BasicTicker(desired_num_ticks=len(colors)),
                     formatter=PrintfTickFormatter(format="%d%%"))

p.add_layout(color_bar, 'right')
hover = p.select(dict(type=HoverTool))
hover.tooltips = [("Year", "@Year"), ("Unviersity", "@University"), ("Admissions", "@Admissions")]
hover.mode = 'mouse'

p.axis.axis_line_color = None
p.axis.major_tick_line_color = None
p.axis.major_label_text_font_size = "7px"
p.axis.major_label_standoff = 0
p.xaxis.major_label_orientation = 1.0
show(p)
```
